## 3. Parameter Definieren

Wir definieren die Batch-Größe für die Anzahl der zu überprüfenden Modelle auf einmal und setzen eine maximale Grenze, wie viele Modelle wir evaluieren möchten.

````python
batch_size = 50
max_models_to_check = 50
german_models = []
````

batch_size: Anzahl der Modelle, die in jeder Anfrage abgerufen werden.
max_models_to_check: Eine Obergrenze, wie viele Modelle wir insgesamt verarbeiten werden.
german_models: Eine Liste, die die Modelle speichert, die die deutsche Sprache unterstützen.

## 4. Funktion zum Überprüfen, ob das Modell ein Sprachmodell ist

Wir definieren eine Hilfsfunktion, um zu überprüfen, ob ein Modell mit Aufgaben der Sprachverarbeitung wie Textgenerierung, Zusammenfassung, Übersetzung oder Klassifizierung in Verbindung steht.

````python
def is_language_model(model_card):
    if "tags" in model_card:
        return any(tag in model_card.tags for tag in ["text-generation", "text2text-generation", "sequence-classification", "summarization", "translation"])
    return False
````

Die Funktion überprüft Tags wie text-generation, sequence-classification usw., um Modelle zu identifizieren, die für Aufgaben der natürlichen Sprachverarbeitung geeignet sind.

5. Hauptfunktion zum Generieren einer Liste verfügbarer deutscher Modelle

In dieser Funktion befragen wir den Hugging Face Hub, verarbeiten die Modelle und filtern sie basierend auf der Unterstützung der deutschen Sprache.

````python
def generate_list_of_available_german_models():
    total_models_processed = 0
    while total_models_processed < max_models_to_check:
        models = list(api.list_models(limit=batch_size, full=False))

        if not models:
            break

        # Iteriere durch die Modelle und überprüfe, ob sie auf Deutsch trainiert sind
        for model_info in models:
            model_name = model_info.id
            try:
                # Lade die Modellkarte
                card = ModelCard.load(model_name)
                card_dict = card.data.to_dict()
                languages = card_dict.get("language")

                # Überprüfe, ob "de" (Deutsch) in der Liste der Sprachen ist
                if languages and "de" in languages:
                    german_models.append(model_name)
                    print(f"Das Modell '{model_name}' unterstützt Deutsch.")
            except Exception as e:
                print(f"Modellkarte für {model_name} konnte nicht geladen werden: {e}")

        total_models_processed += len(models)

        if len(models) < batch_size:
            break

    # Drucke alle Modelle, die Deutsch unterstützen
    print("\nModelle, die Deutsch unterstützen:")
    for model in german_models:
        print(model)

    # Modelle als JSON speichern
    german_models_json = json.dumps(german_models)
    file_path = 'german-models.json'
    with open(file_path, 'w') as json_file:
        json_file.write(german_models_json)
    print(f"Model-Karten-JSON wurde in {file_path} geschrieben")
    return german_models
````

Wichtige Schritte in der Funktion:

    Modellanfragen: Wir verwenden HfApi, um Modelle von Hugging Face abzurufen.
        Wir fragen in Chargen von 50 (festgelegt durch batch_size) an.
        Die Schleife läuft, bis wir max_models_to_check verarbeitet haben.

    Verarbeitung der Modellkarten: Für jedes Modell laden wir die zugehörige ModelCard, die Metadaten über das Modell enthält, einschließlich der unterstützten Sprachen.

    Sprachüberprüfung: Wir prüfen das Feld language der Modellkarte, um zu sehen, ob "de" (Deutsch) enthalten ist. Wenn ja, wird der Modellname zur Liste german_models hinzugefügt.

    Fehlerbehandlung: Wenn die Modellkarte nicht geladen werden kann, wird eine Fehlermeldung ausgegeben.

    Speichern als JSON: Nach der Verarbeitung aller Modelle wird die Liste der Deutsch unterstützenden Modelle in einer Datei german-models.json gespeichert.

## Connector Methods for Model Testing

### load_model_and_tokenizer: Eine zentrale Funktion für NLP-Modelle

Die Funktion load_model_and_tokenizer spielt eine entscheidende Rolle beim Laden vortrainierter Modelle und der zugehörigen Tokenizer aus der Hugging Face Model Hub. Sie ermöglicht eine flexible und modulare Nutzung von Natural Language Processing (NLP)-Modellen und ist besonders wertvoll in Szenarien, in denen verschiedene Modelle getestet oder angewendet werden sollen.

````python
from transformers import (
    AutoConfig,
    AutoModel,
    AutoModelForSeq2SeqLM,
    AutoTokenizer,
    BertModel,
    T5ForConditionalGeneration
)

def load_model_and_tokenizer(model_name_or_object):
    """
    Lädt ein Modell und den entsprechenden Tokenizer basierend auf dem Eingabewert.

    :param model_name_or_object: Entweder ein Modellname (String) oder ein bereits geladenes Modellobjekt.
    :return: Ein Tupel aus (modell, tokenizer).
    """
    if isinstance(model_name_or_object, str):
        # Modell und Tokenizer mit dem Modellnamen laden
        config = AutoConfig.from_pretrained(model_name_or_object)

        if config.is_encoder_decoder:
            model = AutoModelForSeq2SeqLM.from_pretrained(model_name_or_object)
        else:
            model = AutoModel.from_pretrained(model_name_or_object)

        tokenizer = AutoTokenizer.from_pretrained(model_name_or_object)

    elif isinstance(model_name_or_object, (AutoModelForSeq2SeqLM, AutoModel, BertModel, T5ForConditionalGeneration)):
        # Wenn ein Modellobjekt übergeben wird, den zugehörigen Tokenizer laden
        model = model_name_or_object
        tokenizer_name = model.config._name_or_path
        tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)

    else:
        raise ValueError(f"Unsupported model type or input: {type(model_name_or_object)}")

    return model, tokenizer
````

BLEU Score Calculation für Übersetzungsmodelle

Der BLEU (Bilingual Evaluation Understudy) Score ist eine weit verbreitete Metrik zur Bewertung der Qualität von maschinellen Übersetzungen. In diesem Abschnitt präsentieren wir eine Implementierung zur Berechnung des BLEU Scores für Übersetzungen, die mit einem Hugging Face-Modell erstellt wurden.

````python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from model_connection import load_model_and_tokenizer, get_simple_translation

# Datei-Pfade für Datensätze
de_file_path = "datasets/zitate-dewiki-20141024.de"
en_file_path = "datasets/zitate-dewiki-20141024.en"


def categorize_bleu_score(bleu_score):
    if bleu_score > 0.4:
        return "good"
    elif 0.2 <= bleu_score <= 0.4:
        return "average"
    else:
        return "bad"


def calculate_bleu(model_name, max_index=10):
    print(f"BLEU score is being calculated.")

    # Modell und Tokenizer für Hugging Face-Modell laden
    model, tokenizer = load_model_and_tokenizer(model_name)

    # Inhalte aus der ".de"-Datei lesen
    with open(de_file_path, 'r', encoding='utf-8') as de_file:
        de_content = de_file.read()

    # Inhalte aus der ".en"-Datei lesen
    with open(en_file_path, 'r', encoding='utf-8') as en_file:
        en_content = en_file.read()

    # Inhalte in Sätze aufteilen
    de_sentences = de_content.split('\n')
    en_sentences = en_content.split('\n')

    # Gesamten Score initialisieren
    total_bleu_score = 0
    bleu_score_data = {"scores": []}
    count = 0  # Anzahl erfolgreicher BLEU Score-Berechnungen

    # Schleife über die Sätze für Übersetzung und BLEU Score-Berechnung
    for i, (en_sentence, de_sentence) in enumerate(zip(en_sentences, de_sentences)):
        # Hugging Face zur Übersetzung des englischen Satzes ins Deutsche verwenden
        response = get_simple_translation(model, tokenizer, en_sentence, target_language='de')
        reference = de_sentence  # Der deutsche Satz dient als Referenz

        if response and reference:
            # Berechnung des BLEU Scores
            bleu_score = sentence_bleu([reference.split()], response.split(),
                                       smoothing_function=SmoothingFunction().method1,
                                       weights=(1, 0))  # Unigram-BERECHNUNG
            total_bleu_score += bleu_score

            # BLEU Score formatieren
            formatted_bleu_score = "{:.2f}".format(bleu_score)
            score_category = categorize_bleu_score(bleu_score)

            # Ergebnisse speichern
            bleu_score_data["scores"].append({
                "index": i + 1,
                "response": response,
                "reference": reference,
                "bleu_score": formatted_bleu_score,
                "score_category": score_category
            })
            count += 1

        # Schleife beenden, wenn die maximale Anzahl erreicht ist
        if i >= max_index:
            break

    # Durchschnittlichen BLEU Score berechnen
    average_bleu_score = round(total_bleu_score / count, 2)
    formatted_bleu_average_score = "{:.2f}".format(average_bleu_score)
    score_category = categorize_bleu_score(average_bleu_score)

    # Durchschnittswerte zu den Daten hinzufügen
    bleu_avg_data = {
        "BLEU": {
            "count": count,
            "average_bleu_score": formatted_bleu_average_score,
            "score_category": score_category
        }
    }

    print(f"Durchschnittlicher BLEU Score: {average_bleu_average_score}")
    return average_bleu_average_score, score_category
````

Erläuterung des Codes

    Importe: Der Code importiert notwendige Module, darunter sentence_bleu aus der NLTK-Bibliothek, die zur Berechnung des BLEU Scores verwendet wird, sowie Funktionen zur Modell- und Tokenizer-Verwaltung.

    Dateipfade: Die Pfade zu den deutschen und englischen Datendateien werden definiert.

    Kategorisierung: Die Funktion categorize_bleu_score kategorisiert den BLEU Score in "gut", "durchschnittlich" oder "schlecht", basierend auf vordefinierten Schwellenwerten.

    Hauptfunktion:
        Das Modell und der Tokenizer werden geladen.

ROUGE Score Calculation für Textzusammenfassungen

Der ROUGE (Recall-Oriented Understudy for Gisting Evaluation) Score ist eine beliebte Metrik zur Bewertung der Qualität von Textzusammenfassungen. In diesem Abschnitt präsentieren wir eine Implementierung zur Berechnung des ROUGE Scores für Zusammenfassungen, die mit einem Hugging Face-Modell erstellt wurden.

````python
import json
import os
from rouge_score import rouge_scorer
from model_connection import get_answer, load_model_and_tokenizer


def round_rouge_scores(scores):
    return {
        'rouge1': round(scores['rouge1'].fmeasure, 2),
        'rouge2': round(scores['rouge2'].fmeasure, 2),
        'rougeL': round(scores['rougeL'].fmeasure, 2),
    }


def calculate_average_rouge_scores(dataset_points):
    rouge1_scores = [point["rouge1_scores"] for point in dataset_points]
    rouge2_scores = [point["rouge2_scores"] for point in dataset_points]
    rougeL_scores = [point["rougeL_scores"] for point in dataset_points]

    avg_rouge1 = round(sum(rouge1_scores) / len(rouge1_scores), 2)
    avg_rouge2 = round(sum(rouge2_scores) / len(rouge2_scores), 2)
    avg_rougeL = round(sum(rougeL_scores) / len(rougeL_scores), 2)

    return avg_rouge1, avg_rouge2, avg_rougeL


def run_rouge(model_name, tokenizer, max_index=2):
    print("ROUGE score is being calculated.")
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    model, tokenizer = load_model_and_tokenizer(model_name)

    dataset_points = []
    lines = load_data('./datasets/test.json')
    for index, line in enumerate(lines[:max_index]):
        data_point = json.loads(line)

        prompt = f"Fasse den Text auf deutsch zusammen:\n{data_point.get('wiki_sentences')}"

        # Pass both model and tokenizer to get_answer
        response = get_answer(model, tokenizer, prompt)

        reference = data_point.get("klexikon_sentences")
        reference_str = ' '.join(reference)

        scores = scorer.score(reference_str, response)
        rounded_scores = round_rouge_scores(scores)

        dataset_point = {
            "index": index,
            "prompt": prompt,
            "response": response,
            "reference": reference_str,
            "rouge1_scores": rounded_scores['rouge1'],
            "rouge2_scores": rounded_scores['rouge2'],
            "rougeL_scores": rounded_scores['rougeL']
        }

        dataset_points.append(dataset_point)

    avg_rouge1, avg_rouge2, avg_rougeL = calculate_average_rouge_scores(dataset_points)
    return avg_rouge1, avg_rouge2, avg_rougeL


def load_data(json_file_path):
    with open(json_file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
    return lines
````

````python
Erläuterung des Codes

    Importe: Der Code importiert erforderliche Module, darunter rouge_scorer aus der rouge_score-Bibliothek, die zur Berechnung der ROUGE Scores verwendet wird, sowie Funktionen zur Modell- und Tokenizer-Verwaltung.

    Rundung der Scores: Die Funktion round_rouge_scores rundet die ROUGE Scores auf zwei Dezimalstellen für eine bessere Lesbarkeit.

    Durchschnittliche Scores berechnen: Die Funktion calculate_average_rouge_scores berechnet den Durchschnitt der ROUGE Scores aus einer Liste von Datenpunkten.

    Hauptfunktion:
        Die ROUGE Scores werden berechnet, indem ein Modell und Tokenizer geladen werden.
        Die Daten werden aus einer JSON-Datei geladen, die die zu bewertenden Textpunkte enthält.
        Für jeden Datenpunkt wird ein Zusammenfassungsprompt erstellt, und die Antwort des Modells wird aufgezeichnet.
        Die ROUGE Scores werden für die generierte Zusammenfassung und die Referenz berechnet und in einem Dictionary gespeichert.

    Daten laden: Die Funktion load_data liest die Inhalte der angegebenen JSON-Datei und gibt sie als Liste zurück.
````

````python
model_name = "t5-base"  # Beispielmodell
model, tokenizer = load_model_and_tokenizer(model_name)
avg_rouge1, avg_rouge2, avg_rougeL = run_rouge(model_name, tokenizer)
print(f"Durchschnittliche ROUGE Scores - ROUGE-1: {avg_rouge1}, ROUGE-2: {avg_rouge2}, ROUGE-L: {avg_rougeL}")
````

Hate Speech Erkennung mit Hugging Face

In dieser Implementierung wird ein Modell zur Erkennung von Hate Speech verwendet, um Kommentare aus einem Datensatz zu analysieren. Die Ergebnisse werden anschließend ausgewertet und kategorisiert.

````python
import json
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# Dataset path
dataset_path = "./datasets/hate_speech_germeval21_ds.json"


def read_data_from_json(file_path):
    with open(file_path, "r") as f:
        data = json.load(f)
    return data


def categorize_hate_speech_detection(precision_percentage, high_point=80, low_point=40):
    if precision_percentage > high_point:
        return "good"
    elif low_point <= precision_percentage <= high_point:
        return "average"
    else:
        return "bad"


# Funktion zum Laden von Modell und Tokenizer
def load_model_and_tokenizer(model_name):
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    return model, tokenizer


# Funktion zur Vorhersage mit einem Hugging Face Modell
def predict_with_model(model, tokenizer, comment):
    inputs = tokenizer(comment, return_tensors="pt")
    outputs = model(**inputs)
    logits = outputs.logits
    prediction = logits.argmax(dim=-1).item()  # Angenommene binäre Klassifikation, 0 oder 1
    return prediction


# Hauptfunktion zur Hate Speech Erkennung
def run_hate_speech(model_name, max_index=300):
    print(f"Hate speech detection is running.")

    # Lade Modell und Tokenizer von Hugging Face
    model, tokenizer = load_model_and_tokenizer(model_name)

    # Zähle gültige Kommentare und korrekte Antworten
    len_of_valid_comment = 0
    correct_response_count = 0

    # Lade den Datensatz
    dataset = read_data_from_json(dataset_path)

    # Durchlaufe den Datensatz
    for i, row in enumerate(dataset):
        # Extrahiere Kommentar und das markierte Hate Speech-Wert (0 oder 1)
        comment = row['comment_text']
        sub_toxic = row['Sub1_Toxic']

        # Vorhersage mit dem Hugging Face Modell
        model_prediction = predict_with_model(model, tokenizer, comment)

        # Überprüfe, ob die Vorhersage korrekt ist
        if model_prediction == sub_toxic:
            correct_response_count += 1

        len_of_valid_comment += 1

        # Schleife abbrechen, wenn die maximale Anzahl erreicht ist
        if i >= max_index:
            break

    # Berechne den Prozentsatz der korrekten Antworten
    precision_percentage = round(correct_response_count / len_of_valid_comment * 100, 2) if len_of_valid_comment > 0 else 0

    # Rückgabe des Prozentsatzes und der Kategorisierung (gut, durchschnittlich, schlecht)
    return precision_percentage, categorize_hate_speech_detection(precision_percentage)
````

Erläuterung des Codes

    Importe: Der Code verwendet die transformers-Bibliothek von Hugging Face, um ein vortrainiertes Modell für die Klassifikation von Hate Speech zu laden.

    Datensatz lesen: Die Funktion read_data_from_json lädt die Daten aus einer JSON-Datei.

    Kategorisierung: Die Funktion categorize_hate_speech_detection bewertet die Genauigkeit der Vorhersagen und kategorisiert sie als "gut", "durchschnittlich" oder "schlecht".

    Modell laden: Die Funktion load_model_and_tokenizer lädt das Modell und den Tokenizer basierend auf dem übergebenen Modellnamen.

    Vorhersage: predict_with_model nimmt einen Kommentar und gibt die Klassifikation zurück (0 oder 1), basierend auf dem Output des Modells.

    Hauptfunktion:
        Die Funktion run_hate_speech steuert den gesamten Prozess. Sie lädt das Modell, iteriert über die Kommentare im Datensatz und nutzt die Vorhersagefunktion.
        Am Ende berechnet sie den Prozentsatz korrekter Vorhersagen und kategorisiert die Ergebnisse.

````python

model_name = "t5-base"
model, tokenizer = load_model_and_tokenizer(model_name)
precision_percentage, category = run_hate_speech(model_name, tokenizer)  # Pass the tokenizer here
print(f"Präzisionsprozent: {precision_percentage}%, Kategorie: {category}")

````

# Word Connotation Recognition

Dieser Code analysiert die Konnotation von Wörtern mithilfe eines vortrainierten Modells von Hugging Face.


````python
import json
import Levenshtein
from transformers import AutoModelForSequenceClassification, T5ForConditionalGeneration

from model_connection import get_answer

prompt_template = "Categorize which sentiment the example word contains: "
prompt_additional_instructions = "Respond in all lower caps and in a single word."
ds_json_file_path = "datasets/word_connotation_ds.json"


def load_data(ds_json_file_path):
    with open(ds_json_file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    return data.get("Wörter", [])


def generate_prompt(prompt_template, word):
    return f"{prompt_template} '{word}', 'very positive', 'positive', 'neutral', 'negative' or 'very negative'.{prompt_additional_instructions}"


def create_json_file(data, output_file_path):
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        json.dump(data, output_file, ensure_ascii=False, indent=2)


def check_sentiment_match_exact(response_sentiment, reference_sentiment, allowed_distance=1):
    # If response_sentiment is a list, consider the first element
    if isinstance(response_sentiment, list):
        response_sentiment_lower = response_sentiment[0].lower()
    else:
        response_sentiment_lower = str(response_sentiment).lower()

    # Convert reference_sentiment to lowercase for case-insensitive comparison
    reference_sentiment_lower = reference_sentiment.lower()

    # Check if the response sentiment exactly matches the reference sentiment
    exact_match = response_sentiment_lower == reference_sentiment_lower

    # Check if the Levenshtein distance is within the allowed threshold
    distance = Levenshtein.distance(response_sentiment_lower, reference_sentiment_lower)
    within_distance_threshold = distance <= allowed_distance

    # Consider it a match if either the exact match or within_distance_threshold is true
    return exact_match or within_distance_threshold


def check_sentiment_match_in_category(response_sentiment, reference_sentiment):
    # Define sentiment categories
    negative_categories = ['negative', 'very negative']
    positive_categories = ['neutral', 'positive', 'very positive']
    neutral_categories = ['neutral']

    # If response_sentiment is a list, consider the first element
    if isinstance(response_sentiment, list):
        response_sentiment_lower = response_sentiment[0].lower()
    else:
        response_sentiment_lower = str(response_sentiment).lower()

    # Convert reference_sentiment to lowercase for case-insensitive comparison
    reference_sentiment_lower = reference_sentiment.lower()

    # Check if the model recognized the general sentiment correctly
    if reference_sentiment_lower in negative_categories:
        return response_sentiment_lower in negative_categories
    elif reference_sentiment_lower in positive_categories:
        return response_sentiment_lower in positive_categories
    elif reference_sentiment_lower in neutral_categories:
        return response_sentiment_lower in neutral_categories
    else:
        return False


def run_word_connotation_recognition(model, tokenizer):
    # Check model compatibility
    if not isinstance(model, (AutoModelForSequenceClassification, T5ForConditionalGeneration)):
        return None  # or raise an exception
    print(f"Word Connotation Recognition is running.")
    words_data = load_data(ds_json_file_path)
    results = []
    correct_exact_matches = 0
    correct_category_matches = 0

    for word_info in words_data:
        word = word_info.get("Wort", "")
        reference_sentiment = word_info.get("Sentiment", "")

        # Generate prompt for the Hugging Face model
        prompt = generate_prompt(prompt_template, word)

        # Tokenize input prompt using the Hugging Face tokenizer
        inputs = tokenizer(prompt, return_tensors="pt", truncation=True)

        # Generate response using the Hugging Face model
        outputs = model.generate(**inputs, max_new_tokens=50)

        # Decode the model's output
        sentiment_response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

        # Check if the sentiment matches the reference sentiment exactly or in category
        exact_match = check_sentiment_match_exact(sentiment_response, reference_sentiment)
        category_match = check_sentiment_match_in_category(sentiment_response, reference_sentiment)

        # Collect results
        result_entry = {
            'word': word,
            'prompt': prompt,
            'response': sentiment_response,
            'reference_sentiment': reference_sentiment,
            'exact_match': exact_match,
            'category_match': category_match
        }
        results.append(result_entry)

        if exact_match:
            correct_exact_matches += 1
        if category_match:
            correct_category_matches += 1

    total_entries = len(words_data)
    percentage_exact_matches = round((correct_exact_matches / total_entries), 2)
    percentage_category_matches = round((correct_category_matches / total_entries), 2)

    def categorize_results(percentage_matches, threshold_good=85, threshold_bad=65):
        if percentage_matches >= threshold_good:
            return "good"
        elif percentage_matches >= threshold_bad:
            return "average"
        else:
            return "bad"

    result_category_exact_match = categorize_results(percentage_exact_matches, 0.7, 0.5)
    result_category_category_match = categorize_results(percentage_category_matches)

    return percentage_category_matches
````

## Modell und Tokenizer laden und ausführen

````python
model_name = "t5-base"
model, tokenizer = load_model_and_tokenizer(model_name)
percentage_category_matches = run_word_connotation_recognition(model, tokenizer)
print(f"Percentage of category matches: {percentage_category_matches * 100:.2f}%")
````
